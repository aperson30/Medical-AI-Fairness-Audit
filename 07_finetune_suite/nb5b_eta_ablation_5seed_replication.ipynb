```python
import os
os.environ["PYTORCH_ALLOC_CONF"] = "expandable_segments:True"

import torch
import torch.nn as nn
import torch.optim as optim
from torch.cuda.amp import autocast, GradScaler
from torch.utils.data import DataLoader, Dataset
import numpy as np, pandas as pd, math, warnings
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.metrics import roc_auc_score, accuracy_score
from sklearn.preprocessing import LabelEncoder
from transformers import CLIPModel, CLIPProcessor, CLIPVisionModel
warnings.filterwarnings('ignore')

# ── Constants ─────────────────────────────────────────────────
BASE_SEED    = 42
SEEDS        = [42, 0, 1, 7, 99]
ETA_VALUES   = [0.001, 0.01, 0.1, 1.0]
N_EPOCHS_FT  = 5
BATCH_FT     = 16
LR_FT        = 1e-4
WD_FT        = 1e-2
N_CLASSES    = 3
N_GROUPS     = 9
COLLAPSE_THR = 0.01
MAD_I_GATE   = 0.02
N_UNFREEZE   = 4

NC_NG_FITZ     = 0.094
N_MIN_FITZ     = 130
MU_COSINE_FITZ = 0.712

def compute_mad_risk_new(mu_cosine, n_minority_train, nc_ng):
    if nc_ng < MAD_I_GATE:
        return 'MAD-I', None
    return 'MAD-G', (1.0 - mu_cosine) / math.log(1.0 + n_minority_train)

stage_fitz, mad_score_fitz = compute_mad_risk_new(MU_COSINE_FITZ, N_MIN_FITZ, NC_NG_FITZ)
print(f"Fitzpatrick17k MAD Risk Score (new formula):")
print(f"  nc/Ng={NC_NG_FITZ}, n_min={N_MIN_FITZ}, mu_cosine={MU_COSINE_FITZ}")
print(f"  Stage={stage_fitz}")
print(f"  MAD Risk = (1 - {MU_COSINE_FITZ}) / log(1 + {N_MIN_FITZ}) = {mad_score_fitz:.4f}")
print(f"  This score is CONSTANT across all η values.")
print()

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {device}")

# ── Part A ────────────────────────────────────────────────────
print("=== PART A: Frozen Probe Score Update (Table 7) ===")
FROZEN_RESULTS = [
    {'eta': 0.001, 'benign_acc': 0.000, 'demo_auc': 0.640, 'sgg': 0.180, 'collapse_n': 0, 'final_min_wt': 0.091},
    {'eta': 0.01,  'benign_acc': 0.000, 'demo_auc': 0.642, 'sgg': 0.178, 'collapse_n': 0, 'final_min_wt': 0.012},
    {'eta': 0.1,   'benign_acc': 0.076, 'demo_auc': 0.636, 'sgg': 0.183, 'collapse_n': 5, 'final_min_wt': 0.000},
    {'eta': 1.0,   'benign_acc': 0.259, 'demo_auc': 0.578, 'sgg': 0.241, 'collapse_n': 5, 'final_min_wt': 0.000},
]
print(f"MAD Risk Score (new formula): {mad_score_fitz:.4f} [constant across all η]")
print()
print(f"{'η':>8} {'MAD Risk':>10} {'Benign Acc':>12} {'Demo AUC':>10} {'SGG':>8} {'Collapse':>10} {'Final Min Wt':>14}")
for r in FROZEN_RESULTS:
    print(f"{r['eta']:>8.3f} {mad_score_fitz:>10.4f} {r['benign_acc']:>12.3f} "
          f"{r['demo_auc']:>10.3f} {r['sgg']:>8.3f} "
          f"{r['collapse_n']:>8}/5 {r['final_min_wt']:>14.4f}")
print()
print("KEY INSIGHT: η=0.001 and η=0.01 — no weight collapse, but benign_acc=0.000")
print("  Confirms nc/Ng as primary driver, not optimizer step size.")

pd.DataFrame([{
    'eta': r['eta'], 'mad_risk_new': round(mad_score_fitz, 4),
    'mu_cosine': MU_COSINE_FITZ, 'n_min': N_MIN_FITZ,
    'benign_acc_mean': r['benign_acc'], 'demo_auc_mean': r['demo_auc'],
    'sgg_mean': r['sgg'], 'collapse_n_of_5': r['collapse_n'],
    'final_min_wt': r['final_min_wt']
} for r in FROZEN_RESULTS]).to_csv('eta_ablation_frozen_score_update.csv', index=False)
print("Saved: eta_ablation_frozen_score_update.csv")

# ── Dataset ───────────────────────────────────────────────────
_fitz_csv = None
for _root, _, _files in os.walk('/kaggle/input'):
    for _f in _files:
        if _f.endswith('.csv') and 'fitzpatrick' in _f.lower():
            _fitz_csv = os.path.join(_root, _f); break
    if _fitz_csv: break

fitz_csv     = _fitz_csv or '/kaggle/input/fitzpatrick17k/fitzpatrick17k.csv'
fitz_img_dir = os.path.dirname(fitz_csv)
print(f"\nCSV: {fitz_csv}")

df = pd.read_csv(fitz_csv)
df = df[df['fitzpatrick_scale'] > 0]
image_files = {}
for _r, _, _fs in os.walk(fitz_img_dir):
    for _f in _fs:
        if _f.lower().endswith(('.jpg', '.png')):
            image_files[os.path.splitext(_f)[0]] = os.path.join(_r, _f)
df['local_path'] = df['md5hash'].map(image_files)
df = df[df['local_path'].notna()].copy()
df['skin_group'] = df['fitzpatrick_scale'].apply(
    lambda x: 'light' if x <= 2 else ('medium' if x <= 4 else 'dark'))

le = LabelEncoder()
le.fit(['benign', 'malignant', 'non-neoplastic'])
BENIGN_IDX = int(list(le.classes_).index('benign'))
MINORITY_G = 2 * N_CLASSES + BENIGN_IDX

np.random.seed(BASE_SEED)
light_df  = df[df['skin_group'] == 'light'].sample(1000, random_state=BASE_SEED)
medium_df = df[df['skin_group'] == 'medium'].sample(1000, random_state=BASE_SEED)
dark_df   = df[df['skin_group'] == 'dark'].copy()

print("\nLoading CLIP processor (CPU only)...")
clip_proc = CLIPProcessor.from_pretrained("openai/clip-vit-large-patch14")
print("Processor ready.")

class FitzDataset(Dataset):
    def __init__(self, dataframe, processor):
        self.df       = dataframe.reset_index(drop=True)
        self.proc     = processor
        self.skin_map = {'light': 0, 'medium': 1, 'dark': 2}
        self.lmap     = {c: i for i, c in enumerate(le.classes_)}
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        try:
            img = Image.open(row['local_path']).convert('RGB')
        except:
            img = Image.new('RGB', (224, 224), (128, 128, 128))
        pix = self.proc(images=img, return_tensors='pt')['pixel_values'].squeeze(0)
        lbl = self.lmap.get(row['three_partition_label'], 0)
        sk  = self.skin_map.get(row['skin_group'], 0)
        grp = sk * N_CLASSES + lbl
        return pix, torch.tensor(lbl), torch.tensor(grp)


def load_clip_vision_partial():
    """
    Strategy:
      1. Load full CLIPModel in fp16 on CPU → grab proj_weight → delete
      2. Load CLIPVisionModel in fp16 on CPU → freeze all → move frozen layers to GPU as fp16
      3. Cast ONLY the trainable layers (last N_UNFREEZE + post_layernorm) to fp32
         so GradScaler can unscale them (GradScaler requires fp32 params)
      4. proj and head are fp32 from the start
    Frozen fp16 layers act as a cheap feature extractor.
    Trainable fp32 layers + AMP autocast = stable mixed-precision training.
    """
    # Step 1: grab proj weights
    full = CLIPModel.from_pretrained("openai/clip-vit-large-patch14",
                                     torch_dtype=torch.float16)
    proj_weight = full.visual_projection.weight.data.float().clone()  # save as fp32
    out_dim, in_dim = proj_weight.shape   # (768, 1024)
    print(f"    proj shape: ({out_dim}, {in_dim})")
    del full; torch.cuda.empty_cache()

    # Step 2: load vision model in fp16
    vision = CLIPVisionModel.from_pretrained("openai/clip-vit-large-patch14",
                                             torch_dtype=torch.float16)

    # Step 3: freeze everything, then selectively upcast trainable parts to fp32
    for p in vision.parameters():
        p.requires_grad = False

    trainable_modules = (
        list(vision.vision_model.encoder.layers[-N_UNFREEZE:])
        + [vision.vision_model.post_layernorm]
    )
    for mod in trainable_modules:
        mod.float()                          # upcast to fp32
        for p in mod.parameters():
            p.requires_grad = True

    # Gradient checkpointing on encoder (safe on CLIPVisionModel)
    vision.vision_model.encoder.gradient_checkpointing = True

    vision = vision.to(device)

    # Step 4: proj as fp32 Linear
    proj = nn.Linear(in_dim, out_dim, bias=False).to(device)
    proj.weight.data.copy_(proj_weight)
    for p in proj.parameters():
        p.requires_grad = True

    n_tr = (sum(p.numel() for p in vision.parameters() if p.requires_grad)
            + sum(p.numel() for p in proj.parameters()))
    n_tot = (sum(p.numel() for p in vision.parameters())
             + sum(p.numel() for p in proj.parameters()))
    print(f"    Trainable: {n_tr:,} / {n_tot:,} ({100*n_tr/n_tot:.1f}%)")
    return vision, proj, in_dim, out_dim


def get_image_feats(vision, proj, pix):
    """
    fp16 frozen layers → pooler_output (fp16 inside autocast) →
    post_layernorm (fp32, autocast upcasts automatically) →
    proj (fp32) → L2-norm → fp32 output
    """
    out   = vision(pixel_values=pix)
    feats = out.pooler_output.float()           # ensure fp32 before proj
    feats = proj(feats)                          # fp32
    feats = feats / feats.norm(dim=-1, keepdim=True)
    return feats


def run_dro_finetuning(train_df, test_df, eta, seed):
    torch.manual_seed(seed)
    np.random.seed(seed)

    vision, proj, in_dim, out_dim = load_clip_vision_partial()
    head    = nn.Linear(out_dim, N_CLASSES).to(device)   # fp32 by default
    scaler  = GradScaler()
    loss_fn = nn.CrossEntropyLoss(reduction='none')

    # Optimizer only sees fp32 params — GradScaler is happy
    opt = optim.AdamW(
        [p for p in vision.parameters() if p.requires_grad]
        + list(proj.parameters())
        + list(head.parameters()),
        lr=LR_FT, weight_decay=WD_FT
    )

    train_ds = FitzDataset(train_df, clip_proc)
    test_ds  = FitzDataset(test_df,  clip_proc)
    train_dl = DataLoader(train_ds, batch_size=BATCH_FT, shuffle=True,
                          num_workers=2, pin_memory=True)
    test_dl  = DataLoader(test_ds,  batch_size=BATCH_FT, shuffle=False,
                          num_workers=2, pin_memory=True)

    q = torch.ones(N_GROUPS, device=device) / N_GROUPS
    final_min_wt = 1.0 / N_GROUPS

    for epoch in range(N_EPOCHS_FT):
        vision.train(); proj.train(); head.train()
        for pix, lbls, grps in train_dl:
            pix  = pix.to(device, non_blocking=True)
            lbls = lbls.to(device, non_blocking=True)
            grps = grps.to(device, non_blocking=True)

            with autocast():
                feats  = get_image_feats(vision, proj, pix)
                logits = head(feats)
                losses = loss_fn(logits, lbls)

            group_losses = torch.zeros(N_GROUPS, device=device)
            for g in range(N_GROUPS):
                mask = (grps == g)
                if mask.sum() > 0:
                    group_losses[g] = losses[mask].mean()

            weighted = (q * group_losses).sum()
            opt.zero_grad()
            scaler.scale(weighted).backward()
            scaler.step(opt)
            scaler.update()

            with torch.no_grad():
                q *= torch.exp(eta * group_losses.detach())
                q /= q.sum()

        final_min_wt = q[MINORITY_G].item()

    # Evaluate
    vision.eval(); proj.eval(); head.eval()
    all_preds, all_labels, all_probs = [], [], []
    with torch.no_grad():
        for pix, lbls, _ in test_dl:
            pix = pix.to(device, non_blocking=True)
            with autocast():
                feats  = get_image_feats(vision, proj, pix)
                logits = head(feats)
            preds = logits.argmax(dim=1).cpu().numpy()
            probs = torch.softmax(logits.float().cpu(), dim=1).numpy()
            all_preds.extend(preds)
            all_labels.extend(lbls.numpy())
            all_probs.extend(probs)

    all_preds  = np.array(all_preds)
    all_labels = np.array(all_labels)
    all_probs  = np.array(all_probs)
    bm    = (all_labels == BENIGN_IDX)
    b_acc = accuracy_score(all_labels[bm], all_preds[bm]) if bm.sum() > 0 else 0.0
    try:
        d_auc = roc_auc_score(all_labels, all_probs, multi_class='ovr', average='macro')
    except:
        d_auc = float('nan')
    collapse = final_min_wt < COLLAPSE_THR

    del vision, proj, head, scaler, opt
    torch.cuda.empty_cache()
    return b_acc, d_auc, final_min_wt, collapse


# ── Part B ────────────────────────────────────────────────────
print("\n=== PART B: Fine-Tuning η Ablation (5 seeds — Table 6 replication) ===")

train_df = pd.concat([light_df, medium_df]).reset_index(drop=True)
test_df  = dark_df.reset_index(drop=True)

results_ft = []
out_csv_ft = 'eta_ablation_finetuning_5seed.csv'

for eta in ETA_VALUES:
    print(f"\nη = {eta}")
    for seed in SEEDS:
        b_acc, d_auc, min_wt, collapse = run_dro_finetuning(train_df, test_df, eta, seed)
        row = {
            'eta': eta, 'seed': seed,
            'mad_risk':   round(mad_score_fitz, 4),
            'benign_acc': round(b_acc, 4),
            'demo_auc':   round(d_auc, 4),
            'min_wt':     round(min_wt, 4),
            'collapse':   collapse
        }
        results_ft.append(row)
        print(f"  seed={seed}: benign_acc={b_acc:.3f} demo_auc={d_auc:.3f} "
              f"min_wt={min_wt:.4f} collapse={collapse}")
        pd.DataFrame(results_ft).to_csv(out_csv_ft, index=False)
    print(f"  Saved {out_csv_ft}")

df_ft = pd.DataFrame(results_ft)

# ── Plot ──────────────────────────────────────────────────────
eta_summary = df_ft.groupby('eta').agg(
    benign_mean=('benign_acc', 'mean'),
    benign_std=('benign_acc',  'std'),
    auc_mean=('demo_auc',      'mean'),
    collapse_n=('collapse',    'sum')
).reset_index()

fig, ax = plt.subplots(figsize=(8, 5))
ax.errorbar(eta_summary['eta'], eta_summary['benign_mean'],
            yerr=eta_summary['benign_std'], fmt='o-',
            color='#2980b9', capsize=5, linewidth=2, markersize=8,
            label='Benign acc (5 seeds)')
for _, row in eta_summary.iterrows():
    ax.annotate(f"collapse={int(row['collapse_n'])}/5",
                (row['eta'], row['benign_mean']),
                textcoords='offset points', xytext=(8, 5), fontsize=9)
ax.set_xscale('log')
ax.set_xlabel('DRO η (log scale)', fontsize=12)
ax.set_ylabel('Dark-skin benign accuracy', fontsize=12)
ax.set_title('Fine-Tuning η Ablation (5 seeds, CLIP ViT-L/14)\n'
             f'MAD Risk Score = {mad_score_fitz:.4f} [constant — pre-intervention]', fontsize=12)
ax.legend(fontsize=11)
ax.axhline(0, color='#e74c3c', linestyle='--', alpha=0.5, label='Collapse floor')
plt.tight_layout()
plt.savefig('eta_ablation_finetuning_curve.png', dpi=150, bbox_inches='tight')
plt.show()
print("Figure saved: eta_ablation_finetuning_curve.png")

# ── LaTeX ─────────────────────────────────────────────────────
print("\n=== LaTeX TABLE ROWS (Table 6 replacement — 5 seeds) ===")
print(f"Note: MAD Risk Score = {mad_score_fitz:.4f} for all η (pre-intervention property)")
print("$\\eta$ & Benign Acc & Demo AUC & Final Min Wt & Collapse (n=5) \\\\")
print("\\hline")
for _, row in eta_summary.iterrows():
    min_wt_vals = df_ft[df_ft['eta'] == row['eta']]['min_wt']
    print(f"{row['eta']:.3f} & {row['benign_mean']:.3f} $\\pm$ {row['benign_std']:.3f} & "
          f"{row['auc_mean']:.3f} & {min_wt_vals.mean():.4f} $\\pm$ {min_wt_vals.std():.4f} & "
          f"{int(row['collapse_n'])}/5 \\\\")

# ── Interpretation flags ──────────────────────────────────────
print("\n=== INTERPRETATION FLAGS ===")
print(f"MAD Risk Score (new formula, constant across η): {mad_score_fitz:.4f}")
print(f"Formula: (1 - {MU_COSINE_FITZ}) / log(1 + {N_MIN_FITZ}) = {mad_score_fitz:.4f}")
print()
print("Fine-tuning η ablation (5-seed replication):")
for _, row in eta_summary.iterrows():
    print(f"  η={row['eta']:.3f}: benign={row['benign_mean']:.3f}±{row['benign_std']:.3f}, "
          f"collapse={int(row['collapse_n'])}/5")
print()
print("ORIGINAL 2-seed estimates (Table 6 in paper):")
orig = [(0.001, 0.245, 0.783, 9.8, 'Partial'),
        (0.01,  0.361, 0.755, 0.0, 'Full'),
        (0.1,   0.269, 0.758, 0.0, 'Full'),
        (1.0,   0.348, 0.759, 0.0, 'Full')]
for o in orig:
    print(f"  η={o[0]:.3f}: benign={o[1]:.3f}, auc={o[2]:.3f}, "
          f"min_wt={o[3]:.1f}%, collapse={o[4]}")
print()
print("Do 5-seed results confirm or contradict 2-seed estimates?")
print("Flag discrepancies as: 'n=5 replication [confirms/revises] n=2 estimate.'")

# ── Full per-seed table ───────────────────────────────────────
print("\n=== FULL PER-SEED FINE-TUNING TABLE ===")
print(f"{'η':>8} {'seed':>6} {'mad_risk':>10} {'benign_acc':>12} {'demo_auc':>10} {'min_wt':>8} {'collapse':>9}")
for _, row in df_ft.iterrows():
    print(f"{row['eta']:>8.3f} {int(row['seed']):>6} {row['mad_risk']:>10.4f} "
          f"{row['benign_acc']:>12.3f} {row['demo_auc']:>10.4f} "
          f"{row['min_wt']:>8.4f} {str(row['collapse']):>9}")

print("\n✓ Complete.")
print("Files: eta_ablation_frozen_score_update.csv, eta_ablation_finetuning_5seed.csv, eta_ablation_finetuning_curve.png")
```
